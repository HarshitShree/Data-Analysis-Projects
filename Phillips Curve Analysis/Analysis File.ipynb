# Hypothesis testing on the data for the US economy

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import pearsonr

data_x = dfusa["Inflation"]
data_y = dfusa['Unemployment']
alpha = 0.05
correlation_coefficient, p_value = pearsonr(data_x, data_y)
df = len(data_x) - 2


critical_t = abs(stats.t.ppf(alpha / 2, df))


t_statistic = (correlation_coefficient * np.sqrt(df)) / np.sqrt(1 - correlation_coefficient**2)

if abs(t_statistic) > critical_t:
    result = "Reject the null hypothesis: There is a significant correlation."
else:
    result = "Fail to reject the null hypothesis: There is no significant correlation."

print(f"Pearson Correlation Coefficient (r): {correlation_coefficient:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical t-value at alpha = {alpha}: {critical_t:.4f}")
print(result)
```

# Phillips Curve with Regression Line

```python
x = [6.737, 6.815, 6.798, 6.83, 7.014, 7.181, 7.279, 7.487, 7.709, 7.77, 7.957, 8.102, 8.36, 8.531, 8.7, 8.625, 8.536, 8.354, 8.384, 8.319, 8.168, 8.095, 8.037, 7.981, 7.915, 7.842, 7.733, 7.65, 6.51, 10.195, 7.713, 7.33]
y = [13.9, 11.8, 6.3, 10.2, 10.2, 9.0, 7.2, 13.2, 4.7, 4.0, 3.8, 4.3, 3.8, 3.8, 4.2, 5.8, 6.4, 8.3, 10.9, 12.0, 8.9, 9.3, 10.9, 6.4, 5.9, 4.9, 3.6, 4.9, 4.8, 6.6, 4.9, 5.9]

# Calculate the coefficients for the regression line
m, b = np.polyfit(x, y, 1)

# Create the scatter plot
plt.scatter(x, y, label='Data')

# Plot the regression line
plt.plot(x, m * np.array(x) + b, color='red', label='Regression Line')

# Add labels and a legend
plt.xlabel('Unemployment')
plt.ylabel('Inflation')
plt.legend()


sns.lineplot(x=dfindia["Unemployment"],y=dfindia["Inflation"])
# Show the plot
plt.show()
```
